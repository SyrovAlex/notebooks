```python
import keras
import numpy as np
print(keras.__version__)
```

# Темы:  
 - загрузка предобученной модели. Определение состава модели;  
 - получение определенного слоя модели (по индексу, по имени);  
 - определение рзамера на входе и на выходе слоя;  
 - получение списка параметров слоя, обучаемых параметров, не обучаемых параметров;
 - получение весов слоя, задание весов слоя
   

# Загрузка предобученной модели.

```python
model=keras.applications.Xception(include_top=True, weights="imagenet")
```

# Определение состава модели

```python
model.summary()
```

```python
for i, layer in enumerate(model.layers):
    print(i, "", layer.name)
```

# Получаем слой по имени

```python
layer=model.get_layer('block12_sepconv2_bn')
```

# Получаем слой по индексу

```python
layer=model.layers[12]
```

# Немного изучим слой (размер данных на входе и выходе слоя)

```python
print(f'Размер на входе {layer.input.shape}')
print(f'Размер на выходе {layer.output.shape}')
```

# Веса слоя

```python
print(layer.weights)
print(layer.trainable_weights)
print(layer.non_trainable_weights)
```

```python
weights=layer.get_weights()
print(weights)
```

# Зададим новые веса для слоя

```python
new_weights=np.zeros((1,1, 1, 64, 128))
print(new_weights.shape)
```

```python
layer.set_weights(new_weights)
```

```python
print(layer.get_weights())
```

```python
weights_np=np.array(weights)
print(weights_np)
```

```python
with open('layer.npy', 'wb') as f:
    np.save(f, weights_np)
```
